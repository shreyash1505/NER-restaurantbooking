### 2. Loading and parsing the training data

```python
with open('./data/train/'+'1_train_BookRestaurant_full.json', encoding="utf8") as data_file:    
    data = json.load(data_file)
```

```python
data
```

```python
########################## parsing logic #######################

final_dataset = []
sentence_index = 0

for sentence in data['BookRestaurant']:
    words = sentence['data']
    temp = ''
    for word in words:
        if( not word['text'].isspace()):
            try:     
                final_dataset.append([sentence_index,word['text'].rstrip(),word['entity']])
            except:
                final_dataset.append([sentence_index,word['text'].rstrip(),'O'])
            
    sentence_index = sentence_index + 1    
print(final_dataset)
```

```python
len(final_dataset)
```

### 3. Loading the data in pandas dataframe

```python
df = pd.DataFrame(final_dataset,columns=['sentence_index','word','tag'])
df.head()
```

```python
df.to_csv('./data/train/train_book_restaurant.csv', index = False,  encoding='utf-8')
```
